### Recursion:

Recursion refers to a programming concept where a function calls itself during its execution.

Recursive ALgorithm:

1. Base case(when the function will terminate).
2. work toward base.
3. Recursive call.

Application:

1. Tree and Graph Traversal
2. Sorting algorithm
3. Divide and conquer
4. Back tracking algorithm
5. Memoization

**_Call Stack:_** Recursive functions use something called “the call stack.” When a program calls a function, that function goes on top of the call stack.

**_stack overflow/ stck space:_** The stack overflow error happens when the program exceeds the call stack space in memory, this is mostly caused by a recursive code. This call stack is located in the virtual memory.

**_recursive tree:_**: A recursive tree is basically a representative form of recursion which depicts how functions are called and returned as a series of events happening consecutively.

**_Virtual memory:_**

1. text: the text also known as code, is the section of memory which contains executable instructions of a program.
2. data: segment that contains global, static and environment variables, etc.
3. heap: the heap stores dynamic data, complex objects and most of your data needed by the program. This is allocated via system call, and is where the memory fragmentation, leaking occurs and where the Garbage Collector takes place.
4. stack: also known as call stack keeps track of the execution of the program. With a very limited size, it stores local variables and local functions. The call stack uses the stack data structure with the LIFO principle (last-in first-out).


#### Example of some common recurion:


````python
## Print numbers from 1 to N without the help of loops.
def func(n):
    if n == 0:
        return
    func(n - 1)
    print(n, end=" ")  # backtracking


func(5)
````

### how this work:

|   n |   stack |  n==0 | pop | `print(n, end=" ")` |
| --: | ------: | ----: | --: | ------------------: |
|   0 | func(0) | false | pop |
|   1 | func(1) |  true | pop |                   1 |
|   2 | func(2) |  true | pop |                   2 |
|   3 | func(3) |  true | pop |                   3 |
|   4 | func(4) |  true | pop |                   4 |
|   5 | func(5) |  true | pop |                   5 |

#### Dry Run Binary Search


````python
from typing import List


def helperFc(nums: List[int], target: int, l: int, r: int) -> int:
    if l > r:
        return -1

    mid = l + ((r - l) >> 1)

    if nums[mid] > target:
        return helperFc(nums, target, l, mid - 1)
    elif nums[mid] < target:
        return helperFc(nums, target, mid + 1, r)
    else:
        return mid


print(helperFc([1, 2, 3], target=3, l=0, r=2))
````

    void merge(vector<int> &arr, 0, 0, 1)
    {
        vector<int> temp;
        int left = low;
        int right = mid + 1;

        while (left <= mid && right <= high)
        {
            if (arr[left] <= arr[right])
            {
                temp.push_back(arr[left]);
                left++;
            }
            else
            {
                temp.push_back(arr[right]);
                right++;
            }
        }

        while (left <= mid)
        {
            temp.push_back(arr[left]);
            left++;
        }
        while (right <= high)
        {
            temp.push_back(arr[right]);
            right++;
        }
        for (int i = low; i <= high; i++)
        {
            arr[i] = temp[i - low];
        }
    }
