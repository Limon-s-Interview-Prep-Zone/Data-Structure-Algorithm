## 1# Prefix Sum:
Prefix Sum is a powerful technique for efficiently solving problems that involve `cumulative sums of a list or subarray`. It preprocesses an array to store cumulative sums so that range sum queries or computations involving contiguous subarrays can be performed in constant time `O(1)`.<br>
`When To use`
1. Range Sum Queries: When you need to repeatedly calculate the sum of elements within a range 
2. Finding Patterns: Problems where you need to count subarrays or contiguous segments satisfying specific properties (e.g., subarray sum equals a target).
3. Frequency or Count Problems: Problems involving cumulative counts or frequency of elements.

### Formula: Given Nums=`[2, 4, 6, 8, 10]`
1. Compute Prefix sum:
$$ \mathcal{prefix[i]=prefix[i-1]+prefix[i] \text{ ; }} for \text{ }i\gt 0$$
$$\text{initialize }\mathcal{prefix[0]=nums[0]}$$
2. Range $Sum[i:j]$ using Prefix sum:
$$ \mathcal{sum[i:j]=prefix[j]-prefix[i-1]} \text{ if } i>0$$

## 5#Sliding Window:
The sliding window technique is a powerful approach used to solve problems involving subarrays, substrings, or sequences.<br>

**Tips for Sliding Window**
1. Track the Window's Boundaries: Maintain two pointers, start (left) and end (right), to define the window.
2. Dynamic Adjustments: Decide when to expand or shrink the window based on the problem constraints.
Optimize Using Maps: For character frequency problems, use hashmaps to track counts within the window.
3. Avoid Nested Loops: Sliding window typically replaces nested loops in problems involving contiguous sequences. This technique is efficient, reducing the time complexity of many brute-force approaches from $O(n^2)\text{ to } O(n)$

## 6#Fast & Slow Pointers Technique
The Fast & Slow Pointers (or Tortoise and Hare) technique is used for problems involving cycles, finding midpoints, or efficiently traversing linked lists and arrays.  It is used to solve problems efficiently with minimal space complexity (O(1)), while typically operating in linear or sublinear time $(O(n), O(log n))$.

| **Problem**                                                                                  | **Description**                                                                                                   |
|----------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------|
| [141. Linked List Cycle](https://leetcode.com/problems/linked-list-cycle/)                   | Determine if a linked list has a cycle using the **Floyd's Tortoise and Hare** (slow and fast pointers) algorithm. |
| [142. Linked List Cycle II](https://leetcode.com/problems/linked-list-cycle-ii/)             | Find the starting node of the cycle in a linked list if a cycle exists.                                         |
| [876. Middle of the Linked List](https://leetcode.com/problems/middle-of-the-linked-list/)   | Find the middle node of a linked list, using the **two-pointer technique**.                                     |


1. **Cycle Detection:**
   1. Useful for detecting cycles in linked lists or arrays.
   2. It avoids additional memory usage (like hash sets) for tracking visited nodes.
   ```algorithm
      1. SET slow, fast = initial value
      2. Detect a Cycle: 
         1. increment slow+=1 and fast+=2
         2. slow==fast break;//cycle detected
      3. Find the Start of the Cycle: slow!=fast
         1. SET slow = intial value
         2. increment slow and fast to 1
   ```
2. **Finding the Middle of a Sequence:** Useful for dividing linked lists or arrays into two halves for further processing, like in merge sort or palindrome checking.
   ```algorithm
      1. SET slow, fast = initial value
      2. while fast and fast.next
         1. increment slow+=1 and fast+=2
      3. return slow
   ```
3. **Checking Palindromes:** Efficiently compares elements from both ends of a sequence.
   ```algorithm
      1. SET slow, fast = initial value
      2. while fast and fast.next
         1. increment slow+=1 and fast+=2
      3. reverse the second half
      4. compare the 1st half and second half
         1. compare each of them values.
   ```
4. **Finding Duplicate Elements:** Identifies duplicates in arrays with constraints, like a range of 1 to n.
5. **Detecting Convergence:** Determines where two sequences or processes meet(sequences, linked lists, or mathematical functions).
      ```algorithm
      1. SET slow, fast = initial value
      2. while fast!slow:
         1. increment slow+=1 and fast+=2
      3. return slow
   ```


## 7#linkedlist in place traversal:
In-place reversal of a LinkedList is a common technique used in algorithm design to reverse the order of nodes without using additional memory.
- `TC`: $O(N)$
- `SC`: $O(1)$

**Examples:** `1→2→3→4→5→6→null, and k=3`
1. Reverse a Linked List:
2. Reverse a LinkedList in Groups:
   1. Traverse the LinkedList while reversing every group of k nodes.
   2. If the number of remaining nodes is less than k, leave them as is.
   ```bash
   function reverseKGroup(head, k):
    if head is null or k is 1:
        return head

    // Step 1: Count the number of nodes in the list
    length = 0
    current = head
    while current is not null:
        length = length + 1
        current = current.next

    // Step 2: Create a dummy node to simplify edge cases
    dummy = new ListNode(0)
    dummy.next = head
    groupPrev = dummy  // Points to the node before the start of the group

    while length >= k:
        // Step 3: Identify the first node of the group to reverse
        groupStart = groupPrev.next
        groupEnd = groupStart
        for i = 1 to k-1:
            groupEnd = groupEnd.next
        
        // Step 4: Reverse the group of size k
        nextGroup = groupEnd.next
        prev = null
        current = groupStart
        while current != nextGroup:
            temp = current.next
            current.next = prev
            prev = current
            current = temp

        // Step 5: Connect the reversed group with the previous part and the next part
        groupPrev.next = prev   // Connect the previous part to the new reversed head
        groupStart.next = nextGroup  // Connect the end of the group to the next part

        // Step 6: Move the groupPrev pointer to the end of the reversed group
        groupPrev = groupStart

        length = length - k  // Decrease the length by k after reversing a group

    return dummy.next  // Return the new head of the list
   ```
```algorithm
// Work through
1. First Group ( 1 → 2 → 3): 
   1. Reverse 1 → 2 → 3 → null to 3 → 2 → 1 3→2→1.
   2. newHead is now 3, and previousGroupTail is 1.
2. Second Group ( 4 → 5 → 6):
   1. Reverse 4 → 5 → 6 → null to 6 → 5 → 4.
   2. Connect 1 → 6, and previousGroupTail becomes 4.
3. Final List: 3 → 2 → 1 → 6 → 5 → 4 → null.
```

## 8#Monotonic Stack:
A Monotonic Stack is a stack that maintains elements in either increasing or decreasing order. It is used in problems involving intervals, ranges, or nearest greater/smaller elements.
- TC: $O(n)$
- SC: $O(n)$

**Pseducode**
```bash
function nextGreaterElement(nums):
    # Initialize an empty stack to store indices
    stack = []

    # Initialize the result array, initially filled with -1
    result = [-1] * length of nums

    # Loop through each element in the input array
    for i from 0 to length of nums - 1:
        # While the stack is not empty and the current element is greater than
        # the element at the index stored at the top of the stack
        while stack is not empty and nums[i] > nums[stack[-1]]:
            # Pop the index from the stack and update the result for that index
            index = stack.pop()
            result[index] = nums[i]

        # Push the current index onto the stack
        stack.push(i)

    # Return the result array containing the next greater element for each element
    return result
```
***Let’s walk through an example to see how the pseudocode works for the input `[4, 5, 2, 10, 8]`.***
```bash
Initialize:
    stack = []
    result = [-1, -1, -1, -1, -1]
Iterate over the array:
    - i = 0, nums[0] = 4:
        - Stack is empty, so push 0 onto the stack.
        - stack = [0]
    - i = 1, nums[1] = 5:
        - nums[1] = 5 is greater than nums[stack[-1]] = nums[0] = 4.
        - Pop 0 from the stack and update result[0] = 5.
        - Push 1 onto the stack.
        - stack = [1], result = [5, -1, -1, -1, -1]
    - i = 2, nums[2] = 2:
        - nums[2] = 2 is less than nums[stack[-1]] = nums[1] = 5, so just push 2 onto the stack.
        - stack = [1, 2]
    - i = 3, nums[3] = 10:
        - nums[3] = 10 is greater than nums[stack[-1]] = nums[2] = 2.
        - Pop 2 from the stack and update result[2] = 10.
        - nums[3] = 10 is also greater than nums[stack[-1]] = nums[1] = 5.
        - Pop 1 from the stack and update result[1] = 10.
        - Push 3 onto the stack.
        - stack = [3], result = [5, 10, 10, -1, -1]
    - i = 4, nums[4] = 8:
        - nums[4] = 8 is less than nums[stack[-1]] = nums[3] = 10, so just push 4 onto the stack.
        - stack = [3, 4]
Finish:
    - The remaining elements in the stack have no next greater elements, so they remain -1 in the result.
Return result:
    - result = [5, 10, 10, -1, -1]
```

## 9#Top ‘K’ Elements:
The Top 'K' Elements technique is widely used to solve problems that require finding the K largest, smallest, or most frequent elements in an array, stream, or dataset. This technique typically leverages data structures like heaps, priority queues, or quickselect for efficiency.

### Approaches:
1. `Min-Heap / Max-Heap`
   1. Use a `min-heap` to maintain the top K largest elements (smallest at the root).
   2. Use a `max-heap` for the top K smallest elements (largest at the root).
2. `Quickselect (Partitioning)`
   1. Based on the `QuickSort algorithm`.
   2. Finds the Kth largest or smallest element without `fully sorting the array`.
   3. Average time complexity: $O(n)$.
3. `Sorting:`
   1. Sort the array and pick the first or last K elements.
   2. Simpler but less efficient $(O(nlogn))$ compared to heaps or quickselect.

### Heapify:
Heap is a binary tree that satisfies the heap property, but instead of ensuring that each parent node's value is smaller than its children (like in a Min-Heap), in a Max-Heap, each parent node's value is greater than or equal to the values of its children. Heap is often represented as an array, where:
```bash
The root is at index 0.
For a node at index i:
    1. The left child is at index 2i + 1.
    2. The right child is at index 2i + 2.
    3. The parent node is at index (i - 1) // 2.
```

#### Insertion: `(O(logn))`
```bash
Insert(Heap, value):
    Add value at the end of the heap
    i = size of heap - 1
    while i > 0 and Heap[parent(i)] < Heap[i]:
        Swap Heap[i] with Heap[parent(i)]
        i = parent(i)

parent(i):
    return (i - 1) // 2
```
#### Remove:`(O(logn))`
```bash
DeleteMax(Heap):
    if Heap is empty:
        return "Heap is empty"
    
    maxValue = Heap[0]  // The root is the maximum element
    Heap[0] = Heap[size of Heap - 1]  // Replace root with last element
    Remove the last element from the heap
    MaxHeapify(Heap, 0)  // Restore heap property starting from the root
    return maxValue

MaxHeapify(Heap, i):
    left = 2 * i + 1  // Left child index
    right = 2 * i + 2  // Right child index
    largest = i
    
    if left < size of Heap and Heap[left] > Heap[largest]:
        largest = left
    
    if right < size of Heap and Heap[right] > Heap[largest]:
        largest = right
    
    if largest != i:
        Swap Heap[i] with Heap[largest]
        MaxHeapify(Heap, largest)
```
#### Build array to heap: `O(n)`
```bash
BuildMaxHeap(Heap):
    for i = size of Heap // 2 - 1 down to 0:
        MaxHeapify(Heap, i)
```
***summary***
1. Max-Heap maintains the property where the value of each node is greater than or equal to its children, ensuring that the largest element is always at the root.
2. Min-Heap maintains the property where the value of each node is less than or equal to its children, ensuring that the smallest element is always at the root.
3. Insertion involves adding an element at the end and bubbling it up to restore the heap property.
4. Deletion involves removing the root and replacing it with the last element, then bubbling it down to restore the heap property.
5. Heapify is the process of converting an unsorted array into a valid Max-Heap by starting from the last non-leaf node and bubbling down each node.

## 10#Overlaping Interval Problems:
Overlapping interval problems involve working with ranges or intervals of numbers, often requiring sorting, merging, or checking relationships between them. These problems are common in scheduling, range merging, and similar use cases.

## 11#Modified Binary Search
Modified Binary Search is a variation of the classic binary search algorithm applied to solve problems with non-standard search spaces or additional conditions. It is commonly used for problems involving rotated arrays, searching for boundaries, finding minimums/maximums, or specific values in sorted but tweaked data.

1. ***`Leftmost Occurrence:`*** The leftmost occurrence is the first position where the target value appears in the array. For sorted arrays, this can be efficiently found by modifying binary search: <br>
`Algorithm:`
   * Use binary search to find the target.
   * When the target is found, continue searching the left half to ensure it's the first occurrence.
   * Stop when no more leftward occurrences are found.
2. ***`Rightmost Occurrence:`*** The rightmost occurrence is the last position where the target value appears. Similarly, binary search can be used with slight modifications:<br>
`Algorithm:`
   * Use binary search to find the target.
   * When the target is found, continue searching the right half to ensure it's the last occurrence.
   * Stop when no more rightward occurrences are found.

## 12#Binary Tree traversal:
Binary tree traversal refers to visiting all the nodes of a tree in a systematic way. There are two main categories:
1. `Depth-First Traversal (DFS):` Traverses as deep as possible along a branch before backtracking.
2. `Breadth-First Traversal (BFS):` Traverses level by level from the root.

### DFS
```text
        1
       / \
      2   3
     / \   \
    4   5   6
```
1. Inorder(left-root-right): `[4, 2, 5, 1, 3, 6]`
2. Preorder(root-left-right): `[1, 2, 4, 5, 3, 6]`
3. Postorder(left-right-root): `[4, 5, 2, 6, 3, 1]`

#### Preorder(root-left-right): `[1, 2, 4, 5, 3, 6, 7]`
```text
       1
     /   \
    2     3
   / \   / \
  4   5 6   7
```
1. `Recursive:`
   1. Steps:
    1. Start at the root node 1.
    2. Recursively process the left subtree (2 → 4 → 5).
    3. Then, process the right subtree (3 → 6 → 7).
    ```bash
    <!-- recursive -->
        DFS(node)
            if node is NULL
                return
            visit(node)             // Process the current node
            DFS(node.left)          // Recursively traverse the left subtree
            DFS(node.right)         // Recursively traverse the right subtree
    ```
    `Pre-Order Dry Run`

    | **Call Stack**    | **Node** | **Action**       | **Output**           |
    |--------------------|----------|------------------|-----------------------|
    | PRE_ORDER(1)      | 1        | Visit 1          | 1                     |
    | PRE_ORDER(2)      | 2        | Visit 2          | 1, 2                  |
    | PRE_ORDER(4)      | 4        | Visit 4          | 1, 2, 4               |
    | Return to 2       | NULL     | Backtrack         | 1, 2, 4               |
    | PRE_ORDER(5)      | 5        | Visit 5          | 1, 2, 4, 5            |
    | Return to 1       | NULL     | Backtrack         | 1, 2, 4, 5            |
    | PRE_ORDER(3)      | 3        | Visit 3          | 1, 2, 4, 5, 3         |
    | PRE_ORDER(6)      | 6        | Visit 6          | 1, 2, 4, 5, 3, 6      |
    | Return to 3       | NULL     | Backtrack         | 1, 2, 4, 5, 3, 6      |
    | PRE_ORDER(7)      | 7        | Visit 7          | 1, 2, 4, 5, 3, 6, 7   |


2. `Iterative:`
   1. Steps:
      1. Initialize a stack: stack = [1].
      2. Pop 1, visit it, and push its right and left children (3, 2) onto the stack.
      3. Pop 2, visit it, and push its right and left children (5, 4) onto the stack.
      4. Continue until the stack is empty.
   2. Pseudocode:
        ```bash
        DFS_ITERATIVE(root)
            if root is NULL
                return
            create an empty stack
            push root onto the stack
            while stack is not empty
                node = pop from stack
                visit(node)         // Process the current node
                if node.right is not NULL
                    push(node.right) // Push right child to stack
                if node.left is not NULL
                    push(node.left)  // Push left child to stack
        ```
    3. Dry Run: Preorder: Root → Left → Right
        ```dryrun
        1. stack = [1] → Pop 1, visit 1, push 3, 2.
            - Output: 1
            - stack = [3, 2]
        2. stack = [3, 2] → Pop 2, visit 2, push 5, 4.
            - Output: 1, 2
            - stack = [3, 5, 4]
        3. stack = [3, 5, 4] → Pop 4, visit 4.
            - Output: 1, 2, 4
            - stack = [3, 5]
        4. stack = [3, 5] → Pop 5, visit 5.
            - Output: 1, 2, 4, 5
            - stack = [3]
        5. stack = [3] → Pop 3, visit 3, push 7, 6.
            - Output: 1, 2, 4, 5, 3
            - stack = [7, 6]
        6. stack = [7, 6] → Pop 6, visit 6.
            - Output: 1, 2, 4, 5, 3, 6
            - stack = [7]
        7. stack = [7] → Pop 7, visit 7.
            - Output: 1, 2, 4, 5, 3, 6, 7
            - stack = [] (empty stack, traversal ends).
        Final Output: 1, 2, 4, 5, 3, 6, 7
        ```

2. Postorder(left-right-root): `[4, 5, 2, 6, 3, 1]`
3. Level Order(BFS): `[[1], [2, 3], [4, 5, 6]]`

***`Approaches`***
- Use recursion for cleaner implementations.
- Use stacks or queues for iterative methods.
- Understand the traversal order through visualization.

## 14#Backtracking
Backtracking explores all possible solutions and backtracks when a solution path fails. It is especially useful for problems involving combinations, permutations, and searching solutions in a tree-like structure.

`Key Concepts`
1. Recursive Exploration: Explore decisions one step at a time recursively.
2. Constraint Checking: Abandon invalid paths as soon as constraints are violated.
3. Backtrack: Undo decisions and try alternative paths.
4. Base Case: Define a stopping condition for recursion.

`Tips for Efficient Backtracking`
1. Pruning:
   * Use constraints to eliminate invalid paths early.
   * Example: Skip duplicate elements in combinations/permutations.
2. Memoization:
   * Cache results of overlapping subproblems if possible.
3. Marking Visited Nodes:
   * Modify or use auxiliary structures to track visited states.
4. Understand State Space:
   * Visualize the tree of decisions to optimize exploration.

### 0/1 knapsack

The "0/1" in the name means you can either take an item (1) or leave it (0)—you cannot take a fraction of an item.

**Approaches:**

1. `Define the state:` Let `dp[i][w]` represent the maximum value obtainable by considering the `first i items` with a `weight limit of w`.
2. `Recurrence the state:`
   - `For each item i`:
     - `For each weight w`:
       - Include Item `i`: If `w[i] <= w`, add its value to the best solution for remaining weight. $dp[i][w]=dp[i−1][w−w[i]]+v[i]$
       - Exclude item: Use the solution without including `i`. $dp[i][w]=dp[i−1][w]$
       - $dp[i][w]=max(dp[i−1][w],dp[i−1][w− W[i]]+v[i])$
3. `Base Case:` If there are no items ( 𝑖=0 ) or the capacity is zero ( 𝑤= 0), then:
   - $dp[0][w]=0\text{ and }dp[i][0]=0$
4. `Final Decision:` The result is stored in $dp[N-1][W]$
